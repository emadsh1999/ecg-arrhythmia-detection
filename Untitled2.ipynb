```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
train = pd.read_csv('/content/ecg/mitbih_train.csv', header=None)
test = pd.read_csv('/content/ecg/mitbih_test.csv', header=None)

# Last column is the label
X_train = train.iloc[:, :-1].values
Y_train = train.iloc[:, -1].values
X_test = test.iloc[:, :-1].values
Y_test = test.iloc[:, -1].values

print(f"Training samples: {len(X_train)}")
print(f"Each signal length: {X_train.shape[1]} points")
print(f"Classes: {np.unique(Y_train)}")
```

```python
fig, axes = plt.subplots(5, 1, figsize=(12, 10))

labels = ['Normal', 'Arrhythmia Type 1', 'Arrhythmia Type 2',
          'Arrhythmia Type 3', 'Arrhythmia Type 4']
colors = ['green', 'red', 'orange', 'purple', 'blue']

for i in range(5):
    # Find one example of each class
    idx = np.where(Y_train == i)[0][0]
    axes[i].plot(X_train[idx], color=colors[i])
    axes[i].set_title(labels[i], fontsize=12)
    axes[i].set_ylabel('Amplitude')
    axes[i].grid(True, alpha=0.3)

axes[-1].set_xlabel('Time points')
plt.suptitle('ECG Heartbeat Signals by Type', fontsize=14)
plt.tight_layout()
plt.savefig('ecg_signals.png', dpi=150)
plt.show()
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

# Dataset
class ECGDataset(Dataset):
    def __init__(self, X, Y):
        self.X = torch.tensor(X, dtype=torch.float32).unsqueeze(1)
        self.Y = torch.tensor(Y, dtype=torch.long)
    def __len__(self):
        return len(self.X)
    def __getitem__(self, idx):
        return self.X[idx], self.Y[idx]

train_loader = DataLoader(ECGDataset(X_train, Y_train), batch_size=64, shuffle=True)
test_loader  = DataLoader(ECGDataset(X_test, Y_test), batch_size=64)

# 1D CNN Model
class ECGClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(1, 32, kernel_size=5, padding=2), nn.ReLU(), nn.MaxPool1d(2),
            nn.Conv1d(32, 64, kernel_size=5, padding=2), nn.ReLU(), nn.MaxPool1d(2),
            nn.Conv1d(64, 128, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool1d(2),
            nn.Flatten(),
            nn.Linear(128*23, 128), nn.ReLU(),
            nn.Linear(128, 5)
        )
    def forward(self, x):
        return self.net(x)

model = ECGClassifier().cuda()
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

print("Model ready!")
```

```python
epochs = 10
train_losses = []

for epoch in range(epochs):
    model.train()
    total_loss = 0

    for X_batch, Y_batch in train_loader:
        X_batch = X_batch.cuda()
        Y_batch = Y_batch.cuda()

        pred = model(X_batch)
        loss = loss_fn(pred, Y_batch)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)
    train_losses.append(avg_loss)
    print(f"Epoch {epoch+1}/{epochs} - Loss: {avg_loss:.4f}")

print("\nTraining done! ✅")
```

```python
model.eval()
correct = 0
total = 0

with torch.no_grad():
    for X_batch, Y_batch in test_loader:
        X_batch = X_batch.cuda()
        Y_batch = Y_batch.cuda()

        pred = model(X_batch)
        predicted = torch.argmax(pred, dim=1)

        correct += (predicted == Y_batch).sum().item()
        total += len(Y_batch)

accuracy = 100 * correct / total
print(f"Test Accuracy: {accuracy:.2f}%")
```

```python
labels = ['Normal', 'Arrhythmia 1', 'Arrhythmia 2', 'Arrhythmia 3', 'Arrhythmia 4']
colors = ['green', 'red', 'orange', 'purple', 'blue']

model.eval()
X_sample = torch.tensor(X_test[:5], dtype=torch.float32).unsqueeze(1).cuda()

with torch.no_grad():
    preds = torch.argmax(model(X_sample), dim=1).cpu().numpy()

fig, axes = plt.subplots(5, 1, figsize=(12, 10))

for i in range(5):
    true_label = int(Y_test[i])
    pred_label = int(preds[i])
    correct = "✅" if true_label == pred_label else "❌"

    axes[i].plot(X_test[i], color=colors[pred_label])
    axes[i].set_title(
        f"{correct} Real: {labels[true_label]} | Predicted: {labels[pred_label]}",
        fontsize=11
    )
    axes[i].grid(True, alpha=0.3)

plt.suptitle('ECG Arrhythmia Detection — AI Predictions', fontsize=14)
plt.tight_layout()
plt.savefig('ecg_predictions.png', dpi=150)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(range(1, epochs+1), train_losses, marker='o', color='steelblue')
plt.title('Training Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.tight_layout()
plt.savefig('loss_curve.png', dpi=150)
plt.show()
```
